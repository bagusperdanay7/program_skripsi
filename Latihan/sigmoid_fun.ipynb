$$
F(x) = \frac{1}{(1+e^{-x})}
$$


```python
from scipy.special import expit

# menghitung sigmoid function untuk x = 2.5
expit(2.5)
```

Rumus Matematika:

$ F(x) = \frac{1}{(1+e^{-x})} $

$ F(x) = \frac{1}{(1+e^{-2.5})} $

$ F(x) = \frac{1}{(1+0.082)} $

$ F(x) = 0.924 $

```python
# Menghitung nilai fungsi sigmoid
x = 2.5

expit(x)
```

```python
# Calculate Sigmoid Function for Multiple Values

# mendefinisikan nilai list
values = [-2, -1, 0, 1, 2]

# menghitung fungsi sigmoid untu setiap nilai dalam list
expit(values)
```

```python
# Plot sigmoid function untuk jangkauan nilainya

import matplotlib.pyplot as plt
import numpy as np

# mendefinisikan range dari nilai-nilai x
x_val = np.linspace(-10, 10, 100)

# menghitung nilai sigmoid
sigmoid_x = expit(x_val)

# membuat plot
plt.plot(x_val, sigmoid_x)
plt.xlabel('x')
plt.ylabel('F(x)')

plt.show()
```
